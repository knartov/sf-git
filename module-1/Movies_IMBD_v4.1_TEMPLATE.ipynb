# 11. Какого жанра фильмов больше всего?

```python
# +
answers['11'] = 'Drama'
```

```python
# Первое решение, которое пришло в голову

# эту задачу тоже можно решать разными подходами, попробуй реализовать разные варианты
# если будешь добавлять функцию - выноси ее в предобработку что в начале
listik = []
for x in data['genres']:
    listik.append(str(x))
listik2 = []
for x in listik:
    listik2.append(x.split('|'))

listik3 = []

for x in listik2:
    for y in x:
        listik3.append(y)

dictik = {}

for x in listik3:
    if x not in dictik:
        dictik[x] = 0
    else:
        dictik[x] +=1

genres = pd.Series(dictik)

display(genres.sort_values(ascending=False))
```

ВАРИАНТ 2

```python
# Более короткое решение с созданием дополнительного поля в предподготовке
genres_dict = {}
for line in data.genres_list:
    for genre in line:
        if genre not in genres_dict:
            genres_dict[genre] = 0
        else:
            genres_dict[genre] += 1
print(pd.Series(genres_dict).sort_values(ascending=False))
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
# +
answers['27'] = 'Daniel Radcliffe Rupert Grint'
```

```python
mutual_work = {}

for item in data.cast_list:
    for variant in list(combinations(item, 2)):
        if variant not in mutual_work:
            mutual_work[variant] = 0
        else:
            mutual_work[variant] += 1

data27 = pd.Series(mutual_work)
data27.sort_values(ascending=False)
```
